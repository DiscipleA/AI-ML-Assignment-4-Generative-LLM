````python

# Core libraries
import sys, platform, math, json, os, textwrap
from datetime import datetime

# HF Transformers
from transformers import AutoTokenizer, AutoModelForCausalLM, pipeline

# Utilities & analysis
import torch
import pandas as pd
from IPython.display import display, Markdown  # for highlighted text display
import matplotlib.pyplot as plt

# System info
print("Python:", sys.version)
print("Platform:", platform.platform())
print("CUDA available?", torch.cuda.is_available())
if torch.cuda.is_available():
    print("GPU:", torch.cuda.get_device_name(0))

# Helper functions
def wrap(text, width=100):
    return "\n".join(textwrap.wrap(text, width=width))

def highlight(text):
    # Show as bold blockquote in Jupyter for simple emphasis
    formatted_text = "> **{}**".format(text.replace('\n', ' '))
    display(Markdown(formatted_text))
    # display(Markdown(f"> **{text.replace('\n', '')}**"))

def simple_metrics(text: str):
    words = text.split()
    uniq = set(words)
    return {
        "char_len": len(text),
        "word_len": len(words),
        "unique_word_ratio": round(len(uniq)/max(1, len(words)), 4),
    }
````

## Step 2 Model Loading: Load a pre-trained model and its corresponding tokenizer from the Hugging Face Hub.

````python

# Choose a small, open-source model to keep resource usage low.
# You may switch to a Llama 3.2 variant if you have the hardware and license accepted on HF:
# MODEL_ID = "meta-llama/Llama-3.2-1B-Instruct"
# MODEL_ID = "meta-llama/Llama-3.2-3B-Instruct"

MODEL_ID = "TinyLlama/TinyLlama-1.1B-Chat-v1.0"  # default lightweight choice
DEVICE = 0 if torch.cuda.is_available() else -1

print(f"Loading model: {MODEL_ID}")
tokenizer = AutoTokenizer.from_pretrained(MODEL_ID)
generator = pipeline("text-generation", model=MODEL_ID, tokenizer=tokenizer, device=DEVICE)
print("Model ready.")
````

## Step 3 Core Implementation: Implement the code for your chosen task (Generation, Summarization, and Q&A).

### 3.1 Text Generation — *Sheik Speer's* poem (~250 words)

````python
gen_prompt = (
    "Compose 'Sheik Speer\\'s' poem in ~250 words. "
    "Use vivid imagery, a reflective tone, and a touch of classical cadence."
)

# ---- chat-format the prompt for a chat model ----
messages = [
    {"role": "system", "content": "You are a creative poetic assistant."},
    {"role": "user", "content": gen_prompt}
]
chat_text = tokenizer.apply_chat_template(
    messages, tokenize=False, add_generation_prompt=True
)

# ---- generate (TinyLlama chat) ----
out = generator(
    chat_text,
    do_sample=True,
    temperature=0.7,
    top_p=0.95,
    max_new_tokens=280,              # a bit extra headroom for ~250 words
    return_full_text=False,          # <-- critical: don't echo the prompt
    pad_token_id=tokenizer.eos_token_id
)

gen_out = out[0]["generated_text"].strip()

# ---- your existing display/metrics ----
highlight(gen_out)
gen_metrics = simple_metrics(gen_out)
gen_metrics
````

### 3.2 Summarization — summarize the poem (~50 words)

````python
# --- Summarize the previously generated poem (gen_out) to ~80 words ---

assert isinstance(gen_out, str) and len(gen_out) > 0, "gen_out must contain the poem text."

sum_user_prompt = (
    "Summarize the following poem in ~80 words (75–85 words). "
    "Keep the imagery and reflective, classical tone. Output a single paragraph only.\n\n"
    f"{gen_out}"
)

# Chat-format for a *-Chat model (TinyLlama 1.1B Chat)
sum_messages = [
    {"role": "system", "content": "You are a concise literary summarizer."},
    {"role": "user", "content": sum_user_prompt}
]
sum_chat = tokenizer.apply_chat_template(sum_messages, tokenize=False, add_generation_prompt=True)

sum_out = generator(
    sum_chat,
    do_sample=True,
    temperature=0.7,
    top_p=0.95,
    max_new_tokens=160,          # enough room for ~80 words
    return_full_text=False,      # <- don't echo the prompt
    pad_token_id=tokenizer.eos_token_id
)[0]["generated_text"].strip()

# Optional: light post-trim to ~85 words max (keeps things tidy if model runs long)
words = sum_out.split()
if len(words) > 85:
    sum_out = " ".join(words[:85]).rstrip(",;:—- ") + "."

highlight(sum_out)
sum_metrics = simple_metrics(sum_out)
sum_metrics
````

### 3.3 Q&A — Answer the poem's question: *"to be or not to be?"*

````python
# --- Use the generated poem as context for a Q&A prompt ---

qa_context = gen_out
qa_question = "To be or not to be?"

qa_user_prompt = (
    f"Read the following poem and answer the question in one or two thoughtful, poetic sentences.\n\n"
    f"Poem:\n{qa_context}\n\n"
    f"Question: {qa_question}\nAnswer:"
)

# Format for TinyLlama-Chat model
qa_messages = [
    {"role": "system", "content": "You are a reflective literary analyst who responds with poetic insight."},
    {"role": "user", "content": qa_user_prompt}
]
qa_chat = tokenizer.apply_chat_template(qa_messages, tokenize=False, add_generation_prompt=True)

# Generate response
qa_out = generator(
    qa_chat,
    do_sample=True,
    temperature=0.5,
    top_p=0.95,
    max_new_tokens=160,
    return_full_text=False,       # <- prevents the model from echoing the full prompt
    pad_token_id=tokenizer.eos_token_id
)[0]["generated_text"].strip()

# Optional word-length safety trim
qa_words = qa_out.split()
if len(qa_words) > 50:
    qa_out = " ".join(qa_words[:50]).rstrip(",;:—- ") + "."

# Display both the question and answer
display_text = f"**Question:** {qa_question}\n\n**Answer:** {qa_out}"
highlight(display_text)

qa_metrics = simple_metrics(qa_out)
qa_metrics
````

## Step 4 Parameter Experimentation: Max New Tokens/Max Length: (e.g., 50, 150, 300) to control output length.

````python
def run_qa_experiment(max_tokens_values, base_kwargs=None, seed=42):
    """
    Runs the Q&A generation multiple times while varying max_new_tokens.
    Shows each output and returns a DataFrame with metrics and the text.
    """
    assert isinstance(gen_out, str) and len(gen_out) > 0, "gen_out (poem) must be available."

    # Sampling defaults
    base_kwargs = base_kwargs or {"temperature": 0.7, "top_p": 0.95}

    # Make sampling reproducible (per session)
    if seed is not None:
        torch.manual_seed(seed)
        if torch.cuda.is_available():
            torch.cuda.manual_seed_all(seed)

    qa_context = gen_out
    qa_question = "To be or not to be?"

    rows = []
    for mnt in max_tokens_values:
        # Build a fresh chat-formatted prompt each run
        qa_user_prompt = (
            "Read the following poem and answer the question in one or two poetic, thoughtful sentences. "
            "Keep your answer under 50 words.\n\n"
            f"Poem:\n{qa_context}\n\n"
            f"Question: {qa_question}\nAnswer:"
        )
        qa_messages = [
            {"role": "system", "content": "You are a reflective literary analyst who answers briefly and poetically."},
            {"role": "user", "content": qa_user_prompt}
        ]
        qa_chat = tokenizer.apply_chat_template(qa_messages, tokenize=False, add_generation_prompt=True)

        # Generate
        out = generator(
            qa_chat,
            do_sample=True,
            max_new_tokens=mnt,
            return_full_text=False,                 # don't echo prompt
            pad_token_id=tokenizer.eos_token_id,
            **base_kwargs
        )[0]["generated_text"].strip()

        # Post-trim to stay within 50 words (keeps the constraint consistent)
        words = out.split()
        if len(words) > 50:
            out = " ".join(words[:50]).rstrip(",;:—- ") + "."

        # Compute metrics (your helper) + token count
        metrics = simple_metrics(out)
        token_len = len(tokenizer(out)["input_ids"])
        row = {
            "max_new_tokens": mnt,
            **base_kwargs,
            **metrics,
            "token_len": token_len,
            "output": out
        }
        rows.append(row)

        # Show the run output in the notebook
        print(f"=== max_new_tokens={mnt} ===")
        display_text = f"**Question:** {qa_question}\n\n**Answer:** {out}"
        highlight(display_text)
        print()

    return pd.DataFrame(rows)

# --- Run the experiment with three distinct max_new_tokens values ---
exp_values = [50, 150, 300]
qa_exp_df = run_qa_experiment(exp_values)
qa_exp_df
````

## Step 5 Analysis: — How output length affects quality/coherence/style
We examine two simple, proxy metrics vs. `max_new_tokens`:
- **word_len** (length of the answer) — longer budgets can produce more detail but risk rambling/repetition.
- **unique_word_ratio** — rough proxy for lexical variety (not a perfect quality metric).

Use these plots alongside your qualitative judgement (faithfulness, coherence, relevance).


````python

# Plot word_len vs max_new_tokens
plt.figure()
plt.plot([str(v) for v in qa_exp_df["max_new_tokens"]], qa_exp_df["word_len"], marker="o")
plt.title("Q&A: max_new_tokens vs word_len")
plt.xlabel("max_new_tokens")
plt.ylabel("word_len")
plt.grid(True)
plt.show()

# Plot unique_word_ratio vs max_new_tokens
plt.figure()
plt.plot([str(v) for v in qa_exp_df["max_new_tokens"]], qa_exp_df["unique_word_ratio"], marker="o")
plt.title("Q&A: max_new_tokens vs unique_word_ratio")
plt.xlabel("max_new_tokens")
plt.ylabel("unique_word_ratio")
plt.grid(True)
plt.show()
````
